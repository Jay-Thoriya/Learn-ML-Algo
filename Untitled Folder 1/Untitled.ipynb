```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

# sine ware

```python
x = np.arange(0, math.pi*6,0.05)
```

```python
y = np.sin(x)
```

```python
plt.plot(x,y,color="red",linestyle="dotted")
plt.xlabel("angle")
plt.ylabel("sine")
plt.title('sine ware')
plt.grid()
plt.show()
```


# pie plot

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.axis("equal")
subject=["AI","VQR","FLAT","AZURE","SE"]
attendance=[99,89,84,82,40]
ax.pie(attendance, labels=subject, autopct="%.2f%%")
plt.show()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
subject=["AI","VQR","FLAT","AZURE","SE"]
attendance=[99,89,84,82,40]
ax.bar(subject,attendance,  color="green")
plt.show()
```
